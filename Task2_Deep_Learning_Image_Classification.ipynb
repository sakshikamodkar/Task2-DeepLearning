## Objective
The objective of this task is to build and train a **Deep Learning model**

This notebook demonstrates a complete deep learning workflow.

## Step 1: Import Required Libraries

We import the necessary libraries for:
- Deep learning
- Numerical operations
- Data visualization

```python
import tensorflow as tf
from tensorflow.keras import layers, models

import numpy as np
import matplotlib.pyplot as plt
```

## Step 2: Load the MNIST Dataset

The MNIST dataset consists of handwritten digit images (0–9).
It is a standard dataset used for image classification tasks.

```python
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()
```

## Step 3: Explore the Dataset

We check the shape of the training and testing datasets.

```python
print("Training data shape:", X_train.shape)
print("Testing data shape:", X_test.shape)
```

## Step 4: Data Preprocessing

Image pixel values range from 0 to 255.
We normalize them to the range 0–1 to improve model performance.

```python
X_train = X_train / 255.0
X_test = X_test / 255.0
```

## Step 5: Build the Neural Network Model

We create a Sequential deep learning model with:
- Flatten layer
- Dense hidden layers
- Output layer with 10 classes

```python
model = models.Sequential([
    layers.Flatten(input_shape=(28, 28)),
    layers.Dense(128, activation='relu'),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])
```

## Step 6: Compile the Model

The model is compiled using:
- Optimizer: Adam
- Loss function: Sparse Categorical Crossentropy
- Metric: Accuracy

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

## Step 7: Train the Model

The model is trained on the training data.
We also use validation data to monitor performance.

```python
history = model.fit(
    X_train, y_train,
    epochs=5,
    validation_data=(X_test, y_test)
)
```

## Step 8: Evaluate the Model

We evaluate the trained model on the test dataset.

```python
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print("Test Accuracy:", test_accuracy)
```

## Step 9: Visualize Training Results

We visualize:
- Training vs Validation Accuracy
- Training vs Validation Loss

```python
plt.figure(figsize=(12, 4))

# Accuracy plot
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Model Accuracy')

# Loss plot
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.title('Model Loss')

plt.show()
```
